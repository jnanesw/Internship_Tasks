# FCPS

```python

def deCFEmail(fp):
    try:
        r = int(fp[:2],16)
        email = ''.join([chr(int(fp[i:i+2], 16) ^ r) for i in range(2, len(fp), 2)])
        return email
    except (ValueError):
        pass
        
```

```python
for i in range(0,28):
    url="https://www.fcps.edu/schools-centers?items_per_page=10&keywords=&page="
    url=url+str(i)
    req=requests.get(url)
    req.encoding="utf-8"
    response=BS(req.content,"html.parser")
    extract_info = response.findAll("div",{"class":"view__row"})
    
    
    for j in range(0,len(extract_info)):
        inside_extracted_info = extract_info[j].article
        Actual_info = inside_extracted_info.findAll("div",{"class":"node__section"})
        
        Name_of_College = Actual_info[0].div.a.text.replace("\n","")
        Address_And_phoneNumber_of_College = Actual_info[1].text.replace("\n"," ")

        E_mail=""
        Designation=""
        Name=""
        if(Actual_info[2].a):
            E_mail = Actual_info[2].a.get("href").replace("/cdn-cgi/l/email-protection#","")
            E_mail=deCFEmail(E_mail)
            
            Name=Actual_info[2].a.text
            
            Designation=Actual_info[2].div.div.text
            
        if(E_mail==""):
            E_mail="Not Provided"
            Designation="Not Provided"
            Name="Not Provided"
        
        dict1={
            "Name of the college":Name_of_College,
            "Address and phone Number":Address_And_phoneNumber_of_College,
            "Name":Name,
            "Designation":Designation,
            "E-Mail":E_mail
        }
        coll1.insert_one(dict1)
        
        print(i)
        print("Name of the college: "+Name_of_College)
        print("Address and phone Number: "+Address_And_phoneNumber_of_College)
        print("Name: "+Name)
        print("Designation: "+Designation)
        print("E-Mail: "+E_mail)
        print("\n")
        
```

```python
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
```

```python
driver=Chrome(executable_path="C:\WebDrivers\chromedriver.exe")
driver.maximize_window()
driver.get("https://www3.dadeschools.net/SchoolInformation/home")
driver.implicitly_wait(10)

All_items=driver.find_element(By.XPATH, "//select[contains(@name,'perPage')]/option[contains(@value,'4: 1000000')]")
All_items.click()

tr=driver.find_elements(By.XPATH,"//tr[contains(@class,'ng-star-inserted')]")
j=1
for i in tr:
    
    td=i.find_elements(By.TAG_NAME,"td")
    Name = td[3].text.replace("\n","")
    E_Mail = td[0].find_element(By.TAG_NAME,"a").get_attribute("href").replace("mailto:","")
    Address = td[1].text.replace("\n","")
    Phone_Number = td[2].text
    
    dict2={
        "Name":Name,
        "E-Mail":E_Mail,
        "Address":Address,
        "Phone_Number":Phone_Number
    }
    coll2.insert_one(dict2)
    
    print(j)
    print("Name: "+Name)
    print("E-Mail: "+E_Mail)
    print("Address: "+Address)
    print("Phone_Number: "+Phone_Number)
    print("\n")
    j=j+1
```
